```python
import pandas as pd 
```

```python
# Importing my data
# TODO - get the raw link for the zip and extract to extract and use here. 
DATA_PATH_GENRE = 'https://raw.githubusercontent.com/Spotify-Song-Predict-DS18/Song-Predict/main/data/data_w_genres.csv'
DATA_PATH_GENRE_2 = 'https://raw.githubusercontent.com/Spotify-Song-Predict-DS18/Song-Predict/main/data/data_by_genres.csv'
DATA_PATH_ARTIST = 'https://raw.githubusercontent.com/Spotify-Song-Predict-DS18/Song-Predict/main/data/data_by_artist.csv'
DATA_PATH = 'https://raw.githubusercontent.com/Spotify-Song-Predict-DS18/Song-Predict/main/data/data.csv'
genres = pd.read_csv(DATA_PATH_GENRE)
genre2 = pd.read_csv(DATA_PATH_GENRE_2)
artist = pd.read_csv(DATA_PATH_ARTIST)
control = pd.read_csv(DATA_PATH)
```

```python
#CLEANING DATA: REMOVING '' and [] SYMBOLS

list_of_genres = []
for i in genres['genres']:
    i = i[1:-1].replace("'", "")
    list_of_genres.append(i)
    
genres.drop(['genres'], axis=1, inplace=True)
genres = genres.assign(New_genres=list_of_genres)
genres = genres.rename(columns={'New_genres':'genre', 'artists':'artist'})

#removing empty values from 'genre' column
genres = genres[genres['genre'] != '']
```

```python
#CLEANING "ARTISTS": REMOVING [] and ''
list_of_artists  = []
for k in control['artists']:
    k = k[1:-1].replace("'", "")
    list_of_artists.append(k)
list_of_artists   
control.drop(['artists'], axis=1, inplace=True)
control = control.assign(New_artists=list_of_artists)
control = control.rename(columns={'New_artists':'artist'})
```

```python
#Merging 

result = pd.merge(control, genres[['artist', 'genre']], how='left', on='artist')
result = result[result['genre'].isnull() == False]
```

```python
#Final dataframe. Take this df for further model training
result
```

```python
# Transforming my dataset
# TODO - Dates need to be coherent
# TODO - might have to make it into a function. 

result['release_date'] = pd.to_datetime(result['release_date'])

# Model crashes so taking out all music before the beginning 2020
mask = result['release_date'] >= '2020-1-1'
result = result[mask]
```

```python
# Setting my X and Y values, we want to predict 5 songs that fit 
X = result.drop(['name', 'release_date'], axis=1)
y = result['name']
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

# Setting my X_train and y_train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, 
                                                    random_state=42)
```

```python
# Fitting a baseline model that will return 5 songs to recommend 
encoder = OrdinalEncoder() 

# Setting my transformed Feature Matrix
XT_train = encoder.fit_transform(X_train)
XT_test = encoder.fit_transform(X_test)

# Instantiating my baseline model
model = LogisticRegression()

# Fitting on my transformed data
model.fit(XT_train, y_train)
```

```python
def put_in_pred(song):
    # Take the song name and match it to the location of the song in the y
    # Then take it with the artist and match it with the row that it is on the XT_train
    # Then we would fit the song into the model and it will pop out 5 different songs 
    # are similar in nature 
    
    # Creating the mask that will grab the location of the song name
    mask = (y == song)
    yt = y[mask]
    # returns the index locaiton
    yti = list(yt.index)
    yti = yti[0]
    # Now that we have the location of the song we can now match it where 
    # it would be in the X dataframe
    Xt = X.loc[yti]
    Xt = pd.DataFrame(Xt)
    # This returns the row where the song is actually at
    mask = list(Xt.columns)
    Xt = X.loc[mask]
    # Now I have to encode it 
    Xt = encoder.fit_transform(Xt)
    # Now that we have the list of the item, time to run it through the model 
    pred = model.predict(Xt)
    
    
    return pred
```

```python
#Running a test 
song = 'Circles'
artist = 'Mac Miller'
put_in_pred(song)
```

```python
# Now making a function that will take a prediction and return 4 more
def top_five(song):
    song = put_in_pred(song)
    print(song)
    for i in range (0,4):
        song_to_str = ""
        for ele in song:
            song_to_str += ele
        song = put_in_pred(song_to_str)
        print(song)
```
